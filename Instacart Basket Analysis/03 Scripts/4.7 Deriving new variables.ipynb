# Step 1: Complete “price_label” and “busiest_day” Columns

# Create the price_label column using a user-defined function:

```python
# Importing the necessary libraries
import pandas as pd
import os

# Set path
path = r'C:\Users\Jacques\OneDrive\Documents\Data Analytics course\Data Immersion\Section 4\08 April 2025 Instacart Basket Analysis'

# Load the merged DataFrame
ords_prods_merge = pd.read_pickle(os.path.join(path, '02 Data', 'Prepared Data', 'ords_prods_merge.pkl'))
```

```python
# Create a subset of ords_prods_merge
df = ords_prods_merge[:1000000]
```

```python
def price_label(row):
    if row['prices'] <= 5:
        return 'Low-range product'
    elif (row['prices'] > 5) and (row['prices'] <= 15):
        return 'Mid-range product'
    elif row['prices'] > 15:
        return 'High range'
    else:
        return 'Not enough data'

# Apply the function
df['price_label'] = df.apply(price_label, axis=1)
```

```python
# Create the busiest_day column using a for-loop

result = []

for value in df['order_dow']:
    if value == 0:
        result.append('Busiest day')
    elif value == 4:
        result.append('Least busy')
    else:
        result.append('Regularly busy')

df['busiest_day'] = result
```

# Step 2: Create New Column with Two Busiest and Two Slowest Days

```python
# Find the two busiest and two slowest days:
print(df['order_dow'].value_counts())
```

```python
# Assumptions:
# Two busiest: 0 (Saturday), 1 (Sunday)
# Two slowest: 3 (Tuesday), 4 (Wednesday)
```

```python
# Create new column busiest_days
df.loc[df['order_dow'].isin([0, 1]), 'busiest_days'] = 'Busiest days'
df.loc[df['order_dow'].isin([3, 4]), 'busiest_days'] = 'Slowest days'
df.loc[~df['order_dow'].isin([0, 1, 3, 4]), 'busiest_days'] = 'Regular'
```

# Step 3: Check Values for Accuracy

```python
# Prints the number of times each unique value appears in the 'busiest_days' column of the df DataFrame, including any missing values (NaN).
print(df['busiest_days'].value_counts(dropna=False))
```

# Step 4: Create busiest_period_of_day Column

```python
# Check frequency of order hours
print(df['order_hour_of_day'].value_counts())
```

# Defining the labels
# Most orders: Hours 10, 11, 13
# Fewest orders: 3, 4
# All other hours: Average orders

```python
# Assigning the labels 'Most orders', 'Fewest orders', or 'Average orders' to the 'busiest_period_of_day' column in df based on whether the order_hour_of_day falls into high, low, or average activity time ranges.
df.loc[df['order_hour_of_day'].isin([10, 11, 14]), 'busiest_period_of_day'] = 'Most orders'
df.loc[df['order_hour_of_day'].isin([3, 4]), 'busiest_period_of_day'] = 'Fewest orders'
df.loc[~df['order_hour_of_day'].isin([10, 11, 14, 3, 4]), 'busiest_period_of_day'] = 'Average orders'
```

# Step 5: Print Frequency

```python
# Displays the count of each label in the 'busiest_period_of_day' column of the df DataFrame, including any missing values.
print(df['busiest_period_of_day'].value_counts(dropna=False))
```
